```python
with open('car_model.pkl','rb') as model_file: 
    model = pickle.load(model_file)
    
with open('columns_names.json','r') as col_file: 
    column_names = json.load(col_file)
    
with open('encoded_data.json','r') as encode_file:
    encoded_data = json.load(encode_file)

user_input = np.zeros(len(column_names['Column Names']))

array = np.array(column_names['Column Names'])

Make='Volvo'
Type='Wagon'
Origin='Europe'
DriveTrain='Front'
MSRP=26135
Invoice=24641
EngineSize=1.9
Cylinders=4.0
Horsepower=170
MPG_Highway=29
Weight=2822
Wheelbase=101
Length=180
Millage_City=22

Make_string = 'Make_'+Make
Make_index = np.where(array == Make_string)[0][0]
user_input[Make_index] = 1 

user_input[0] = encoded_data['Type'][Type]
user_input[1] = encoded_data['Origin'][Origin]
user_input[2] = encoded_data['DriveTrain'][DriveTrain]
user_input[3] = MSRP
user_input[4] = Invoice
user_input[5] = EngineSize
user_input[6] = Cylinders
user_input[7] = Horsepower
user_input[8] = MPG_Highway
user_input[9] = Weight
user_input[10] = Wheelbase
user_input[11] = Length

print(f"{user_input=}")
print(len(user_input))

millage = model.predict([user_input])
print(f"Predicted millage = {Millage_City}")
print(f"Actual millage = 22")
```
